The agent starts exploring and looking for possible rewarding buttons.

## 3. New case proposal
We decided to propose a modification of the scene Pyramids.
In our version the agent must press 2 buttons in a defined sequence and then pick up the pyramid reward:

    Red Button > Orange Button > Pyramid Reward
    
We decided to work with tags, as the Raytracing Scripts use them to detect what the agent is seeing.
#### Code:

```python
 void OnCollisionEnter(Collision collision)
    {
        if (collision.gameObject.CompareTag("goal"))
        {
            SetReward(2f);
            Done();
        }

        if (collision.gameObject.CompareTag("switchOff_O")&&buttonCounter==1)
        {
            Debug.Log("ORANGE");
            AddReward(0.5f);
        }

        if (collision.gameObject.CompareTag("switchOff_R") && buttonCounter == 0)
        {
            Debug.Log("RED");
            AddReward(0.5f);
        }

        if (collision.gameObject.CompareTag("switchOff_O") && buttonCounter == 0)
        {
            Debug.Log("ORANGE");
            AddReward(-0.3f);
            Done();
        }


    }
```

### Observation
Here you can see that the agent gets its rewards from overlapping with the different buttons.

But as you can see, we reward negatively the agent if he decides to press the orange button first, to ensure that the button is pressed after the orange button.
